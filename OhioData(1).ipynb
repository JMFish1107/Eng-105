 Secondly, one needs to import the larger public data file that they wish to work with by defining it as pd. ## Excecution

### Filtering by State

There are two ways one could trim such a large file in order to better understand data of differing degreees.  The first is through filtering our data using the operating system's true or false analysis.  In the given example from the "County Health Data", the larger data set is filtered through this specific code by reading whether it is true or not that the "State" column in each row reads "OH" or not.  This helps us glean the data only from the state of Ohio.  

```python
OH_subset = df[df["State"] == "OH"].copy()
```

This line simply takes the command defined in the previous line of code that checks whether or not data is coming from the state of Ohio or not and executes the command.  The data which came back as true is then presented in table form just as the larger dataset that came before it.  

```python
OH_subset
```

### Filtering by state and data type.  

The following code specifies both the rows and columns that one wants to look at by calling the specific rows in integer form with a colon inbetween (in this example the rows called are all those including and between rows 4001 and 4010).  The loc function comes into play when it comes to the columns section that is specified by strings rather than integers.  In this example, the strings "State", "High school graduation", "Unemployment", and "Income inequality" are the only columns specified and will therefore, in spite of the numerous other columns in the original dataset, be the only ones conjured.  

```python
OH_subset.loc[4001:4010,["State", "High school graduation", "Unemployment", "Income inequality"]]

```

This final section defines Final_OH as the edited table that we see above.  We will then take this final object and export it as a CSV file.  

```python
Final_OH = OH_subset.loc[4001:4010,["State", "High school graduation", "Unemployment", "Income inequality"]]
```

```python

Final_OH.to_csv("OH_subset.csv")
```
